```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the crypto_data.csv dataset.
file_path = "crypto_data.csv"
df_coin = pd.read_csv(file_path, index_col = 0)
df_coin.head(10)
df_coin.count()
```

```python
# Keep all the cryptocurrencies that are being traded.
df_coin_trading= df_coin[df_coin["IsTrading"]==True]
df_coin_trading.head()
```

```python
df_coin_trading.count()
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
df_coin_trading["Algorithm"].dropna()
```

```python
df_coin_trading.count()
```

```python
# Remove the "IsTrading" column. 
clean_coin_trading=df_coin_trading.drop(["IsTrading"], axis=1)
clean_coin_trading.head()
```

```python
# Remove rows that have at least 1 null value.
crypto_df= clean_coin_trading.dropna()
```

```python
crypto_df.count()
```

```python
crypto_df
```

```python
# Keep the rows where coins are mined.
crypto_df=crypto_df[crypto_df["TotalCoinsMined"]>0]
```

```python
crypto_df.count()
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
crypto_name_df=crypto_df["CoinName"]
```

```python
crypto_name_df
```

```python
crypto_df=crypto_df.drop(["CoinName"], axis=1)
```

```python
crypto_df.head(10)
```

```python
# Use get_dummies() to create variables for text features.
Algorithm= pd.get_dummies(crypto_df["Algorithm"])
Proof= pd.get_dummies(crypto_df["ProofType"])
```

```python
x = pd.concat([Algorithm,Proof],axis =1)
x
```

```python
X=crypto_df.merge(x,left_index = True,right_index = True)
X=X.drop(['Algorithm','ProofType'],axis = 1)
```

```python
X
```

```python
X_Scaled= StandardScaler().fit_transform(X)
print(X_Scaled[0:1])
```

```python
# Using PCA to reduce dimension to three principal components.
pca =PCA(n_components=3)
crypto_pca=pca.fit_transform(X_Scaled)
```

```python
pcs_df=pd.DataFrame(data= crypto_pca, columns=["PC 1", "PC 2", "PC 3"], index= crypto_df.index)
pcs_df
```

```python
# Create an elbow curve to find the best value for K.
#Find the best value for K
inertia=[]
k=list(range(1,11))
# Calculate the inertia for the range of K values
for i in k:
    km= KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
    
# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Initialize the K-Means model.
model= KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(pcs_df)

# Predict clusters

predictions=model.predict(pcs_df)

predictions
```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
clustered_df = pd.concat([crypto_df,pcs_df],axis =1)

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies. 
clustered_df["CoinName"]=crypto_name_df

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
clustered_df["Class"]=model.labels_

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)


```
